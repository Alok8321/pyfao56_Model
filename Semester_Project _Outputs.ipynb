# Squires Cotton

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Squires cotton Pyfao vs SWB.csv')
```

```python
df.columns
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['E_py'], color='blue', label='Pyfao56')
plt.plot(df['Date'], df['E'], color='green', label='SWB (Spreadsheet)')
plt.title('Comparison of Evaporation: Pyfao56 vs SWB (Spreadsheet)', fontsize ="14", weight='bold')
plt.xticks(df['Date'][::30], rotation=0, fontsize='14',ha='center')
plt.yticks(fontsize='14')
plt.xlabel('Date', weight='bold', fontsize='16')
plt.ylabel('Evaporation (mm/d)', weight='bold', fontsize='16')
plt.legend(fontsize='14')
plt.show()
```

```python
# Calculate RMSD
E = np.sqrt(np.mean((df['E_py'] - df['E'])**2))

print("Root Mean Square Deviation (RMSD):", E)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['ETc_py'], color='blue', label='Pyfao56')
plt.plot(df['Date'], df['ETc'], color='green', label='SWB (Spreadsheet)')
plt.title('Comparison of Crop Evapotranspiration: Pyfao56 vs SWB (Spreadsheet)', fontsize ="14", weight='bold')
plt.xticks(df['Date'][::30], rotation=0, fontsize='14',ha='center')
plt.yticks(fontsize='14')
plt.xlabel('Date', weight='bold', fontsize='16')
plt.ylabel('Crop Evapotranspiration (mm/d)', weight='bold', fontsize='16')
plt.legend(fontsize='14')
plt.show()
```

```python
# Calculate RMSD
ETc = np.sqrt(np.mean((df['ETc_py'] - df['ETc'])**2))

print("Root Mean Square Deviation (RMSD):", ETc)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['T_py'], color='blue', label='Pyfao56')
plt.plot(df['Date'], df['T'], color='green', label='SWB (Spreadsheet)')
plt.title('Comparison of Transpiration: Pyfao56 vs SWB (Spreadsheet)', fontsize ="14", weight='bold')
plt.xticks(df['Date'][::30], rotation=0, fontsize='14',ha='center')
plt.yticks(fontsize='14')
plt.xlabel('Date', weight='bold', fontsize='16')
plt.ylabel('Transpiration (mm/d)', weight='bold', fontsize='16')
plt.legend(fontsize='14')
plt.show()
```

```python
# Calculate RMSD
T = np.sqrt(np.mean((df['T_py'] - df['T'])**2))

print("Root Mean Square Deviation (RMSD):", T)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['DP_py'], color='blue', label='Pyfao56')
plt.plot(df['Date'], df['DP'], color='green', label='SWB (Spreadsheet)')
plt.title('Comparison of Deep Percolation: Pyfao56 vs SWB (Spreadsheet)', fontsize ="14", weight='bold')
plt.xticks(df['Date'][::30], rotation=0, fontsize='14',ha='center')
plt.yticks(fontsize='14')
plt.xlabel('Date', weight='bold', fontsize='16')
plt.ylabel('Deep Percolation (mm)', weight='bold', fontsize='16')
plt.legend(fontsize='14')
plt.show()
```

```python
# Calculate RMSD
DP = np.sqrt(np.mean((df['DP_py'] - df['DP'])**2))

print("Root Mean Square Deviation (RMSD):", DP)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Date'], df['Irrig_py'], color='blue', label='Pyfao56', width=1)
plt.bar(df['Date'], df['Irrig'], color='green', label='SWB (Spreadsheet)', width=1)
plt.title('Comparison of Irrigation: Pyfao56 vs SWB (Spreadsheet)', fontsize=14, weight='bold')
plt.xticks(df['Date'][::30], rotation=0, fontsize=14, ha='center')
plt.yticks(fontsize=14)
plt.xlabel('Date', weight='bold', fontsize=16)
plt.ylabel('Irrigation (mm)', weight='bold', fontsize=16)
plt.legend(fontsize=14)
plt.show()
```

```python
# Calculate RMSD
I = np.sqrt(np.mean((df['Irrig_py'] - df['Irrig'])**2))

print("Root Mean Square Deviation (RMSD):", I)
```
